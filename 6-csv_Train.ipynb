```python
#Kütüphaneleri import ediyoruz
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("fashion_output.csv")
```

```python
#Daha önce fotoğrafların benzerlik matrisini sıraladığımız şekilde sıralama yapıyoruz
data_frame=data_frame.sort_values(by=["ProductId"])
data_frame
```

```python
#Product id'ye ihtiyacımız olmadığını anladığımızdan atıyoruz
data_frame.pop("ProductId")
data_frame
```

```python
#Renk için normalizasyon yapıyoruz 
#En yüksek veri bölü tüm veriler şeklinde 1-0 arası değer alıyor 
data_frame['Colour']=data_frame['Colour']/data_frame['Colour'].max() 
data_frame['Colour']
```

```python
#Product Type için normalizasyon yapıyoruz 
#En yüksek veri bölü tüm veriler şeklinde 1-0 arası değer alıyor 
data_frame['ProductType']=data_frame['ProductType']/data_frame['ProductType'].max() 
data_frame['ProductType']
```

```python
#Veriler arasındaki korelasyona bakıyoruz
data_correlations = data_frame.corr(method='pearson',min_periods=80)
data_correlations
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Burada kosinus benzerlik formulu ile elde edilen oz nitelik vektorlerinin benzerliklerini karsilastiriyoruz 

cosSimilarities = cosine_similarity(data_frame)

# daha sonra da her bir fotograf icin bunlari bir data frame icersine kaydediyoruz

cos_similarities_df = pd.DataFrame(cosSimilarities, columns=fashion_title, index=fashion_title)
cos_similarities_df.head()
#2906x2096lik bir benzerlik data frame'i olusturuldu.
```
