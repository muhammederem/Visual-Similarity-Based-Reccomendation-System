Elimizde birbirinin aynisi olan fotograflar var mi yok mu kontrol edip eger varsa bunlarin olmadigi baska bir klasor olusturuyoruz.

```python
#Elimizde birbirinin aynisi olan fotograflar var mi yok mu kontrol edip eger varsa bunlarin olmadigi baska bir klasor olusturuyoruz.
files = ['images/'+ x for x in os.listdir('images/') if "jpg" in x]
```

```python
#Bunun için benzerlik matrisinden yararlanıldı 
#Eğer benzerlik matrisinde köşegen haricinde 1 veya 1'e çok yakın değer varsa bu fotoğraflar birbirinin aynısı demektir 
#Bu yüzden bu fotoğrafları belirli bir  listeye atadık
#Eğer o fotoğraf daha önce listeye eklendiyse eklemiyoruz.
count= 0
listo = []
for x in range(len(files)):
    for y in range(len(files)):
        if(data_frame.iloc[x,y]>=0.999 and x!=y):
            count+=1
            if(x not in listo and y not in listo):
                listo.append(y)
```

```python
#Bu fotoğrafların files'daki konumları belirlenmiş oldu
listo
```

```python
len(listo)
```

```python
#Daha sonra bu fotoğrafların dosya yollarını bir listede tutuyoruz
liste = []
for x in listo:
    liste.append(files[x])
liste
```

```python
#Daha sonra bunları dosyalarımızdan çıkartıyoruz
print(len(files))
for x in liste:
    files.remove(x)
print(len(files))
```

```python
#Aynisi olan fotograflari teker teker eledikten sonra yeni listemizde olan fotograflari baska bir klasore tasidik.
import glob
import shutil
import os

src_dir = ""
dst_dir = "image_new/"
for file in files:
    for jpgfile in glob.iglob(os.path.join(src_dir, file)):
        shutil.copy(jpgfile, dst_dir)
```
