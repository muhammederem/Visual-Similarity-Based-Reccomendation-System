```python
#Kütüphaneleri import ediyoruz
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Görsel veriden elde edilen benzerlik matrisini çağırıyoruz
visual_sim=pd.read_csv("cos_sim_processed.csv")
visual_sim
```

```python
#Numeric veriden elde edilen benzerlik matrisini çağırıyoruz
vectoral_sim=pd.read_csv("csv_cossimilarity.csv")
vectoral_sim
```

```python
#Yüzdelik değerler verip çarpıyoruz
data = vectoral_sim*0.3
data
```

```python
#Yüzdelik değerler verip çarpıyoruz
data2=visual_sim*0.7
data2
```

```python
#matrisleri toplayabilmek için aynı isimde sütunlara sahip olması gerkiyor bu yüzden isimleri değiştiriyoruz
#Sıralar aynı olsa da isimler farklı olduğundan toplama olmuyordu.
data.columns = data2.columns
data
```

```python
#Son benzerlik matrisini elde ediyoruz
result = data.add(data2, fill_value=0)
result
```

```python
#Sonra bu matrisi kaydediyoruz
result.to_csv("Result_Simularity.csv",index = False,doublequote=False)
```
